# ***Cantidad económica de pedido de varios artículos con limitación de almacén***

Lo primero que debemos hacer es Importar la librería que vamos a ocupar para encontrar el valor de lambda óptimo(scipy.optimize), Qué es la más resaltante para este código, además de las librerías que ya hemos trabajado antes.

```python
import numpy as np
import sympy as sp
import scipy.optimize as opt
```

Creamos arreglos(arrays) para guardar los datos que proporciona el problema

```python
n = 3                             # Número de artículos

K = np.array([10, 5, 15])         # Costo de preparación o pedido (K_i)
D = np.array([2, 4, 4])           # Tasa de demanda (D_i)
h = np.array([0.3, 0.1, 0.2])     # Costo unitario de almacenamiento (H_i)
a = np.array([1, 1, 1])           # Área requerida por unidad de inventario (a_i)

A = 25                            # Área máxima disponible

Q = np.array([0, 0, 0])           # Cantidades de pedido (Q_i)
area_total = np.array([0, 0, 0])  # Area total que va a ocupar el pedido
```

# **PASO 1**

Calcular los valores óptimos no restringidos de las cantidades de pedido

```python
Q = np.sqrt((2 * K * D) / h)
```

# **PASO 2**

Comprobar si los valores óptimos no restringidos satisfacen la restricción de almacenamiento.

```python
area_total = np.sum(a * Q )      # Área total ocupada

if area_total <= A:              # Comparación con la restricción
    cumple_restriccion = 1

else:
    cumple_restriccion = 0
```

# **paso 3**

Una vez verificado SI se cumple o NO la restricción, Pasamos a lo siguiente:

Sí se cumple, nos quedamos con los Q ya calculados.

Pero, si no se cumple, Se tiene que pasar al proceso de optimización del lambda para calcular los nuevos Q.

```python
if cumple_restriccion == 1:
    print(" No se requiere optimizar Lambda")
    print("                              ")
    print(" Los valores correspondientes de las cantidades de pedido son:")
    print("                              ")
    print(" Artículo 1:", Q[0])
    print(" Artículo 2:", Q[1])
    print(" Artículo 3:", Q[2])

else:
    print(" La restricción no se cumplió.")
    print("                              ")
    print(" Se inicia la optimización del lambda.")


    def restriccion_lambda(lambda_val):
        Q_lambda = np.sqrt((2 * K * D) / (h + 2 * lambda_val * a))    # Q ajustado con restricción de almacén
        return np.sum(a * Q_lambda) - A

    lambda_inicial = 0.0                                              # multiplicador de Lagrange (Inicialización)
    lambda_opt = opt.fsolve(restriccion_lambda, lambda_inicial)[0]    # Resolver λ
    Q_opt = np.sqrt((2 * K * D) / (h + 2 * lambda_opt * a))        # Cantidades ya ajustadas a la restricción

    print(" Ya una vez optimizado lambda, tenemos:")
    print("                              ")
    print(" Lambda óptimo:", lambda_opt)
    print("                              ")
    print(" asi")
    print("                              ")
    print(" Los valores correspondientes de las cantidades de pedido son:")
    print("                              ")
    print(" Artículo 1:", Q_opt[0])
    print(" Artículo 2:", Q_opt[1])
    print(" Artículo 3:", Q_opt[2])
```

A diferencia de las iteraciones por incrementos o decrementos, Cómo lo hace el autor del libro en la solución del ejercicio, el uso de la librería "scipy.optimize" permitió reemplazar estos métodos iterativos, por un procedimiento numérico automático, más eficiente y preciso.

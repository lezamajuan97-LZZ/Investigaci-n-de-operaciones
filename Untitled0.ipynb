Siguiente código muestra otra forma de hacer la red y además cómo se devuelve el número de nodos y de arcos

```python
G=nx.DiGraph()
G.add_edges_from([(1, 2), (1, 3), (2,4),(3,4), (2,3)])
nx.draw_networkx(G)
n= G.number_of_nodes()
print("esta red consta de", n,"nodos")
m = G.number_of_edges()
print("esta red consta de", m,"arcos")
```

#Camino más corto entre 2 nodos específicos.

Nuestro problema dice lo siguente.

"Hallar la ruta más corta entre O y T"

lo primero que hacemos es formar nuestra red o grafo.
que para hacerlo mas sensillo ocupamos uno visto en clase.

```python
W=nx.Graph()
W.add_weighted_edges_from([("O","A",2),("O","B",5),("O","C",4),("A","B",2),("A","D",7),("B","C",1),("B","E",3),("B","D",4),("E","D",1),("E","T",7),("D","T",5)])

print("Los vértices de W son {}".format(list(W.nodes))) # Este comando crea una lista de los nodos
print("Las aristas de W son {}".format(list(W.edges)))  # Este comando crea una lista De los arcos
print("Las aristas de W y sus DISTANCIAS son {}".format(list(W.edges(data=True)))) # Este comando también crea una lista de los arcos, además de indica cuál es la longitud de ellos
nx.draw_networkx(W)
```

```python

ruta = nx.shortest_path       (W, source="O", target="T", weight="weight") # Este comando nos devuelve la ruta más corta Entre 2 especificos
dist = nx.shortest_path_length(W, source="O", target="T", weight="weight") # Este comando nos devuelve la distancia de la ruta mas corta

print(f"Ruta más corta: {ruta}")
print(f"Distancia total: {dist}")
```

#Arbol de expancion minima

```python

arbol_minimo = nx.minimum_spanning_tree(W, weight='weight')
print(arbol_minimo)
```

# Máximo flujo

Para ver con funcionar el código del fujo máximo También utilizaremos un ejemplo vista en clase

" Hallar el flujo máximo de A a F"

primero creamos nuestra red

```python
J=nx.DiGraph()
J.add_edge("A","B",capacity=9)
J.add_edge("A","C",capacity=7)
J.add_edge("B","D",capacity=7)
J.add_edge("B","E",capacity=2)
J.add_edge("C","D",capacity=4)
J.add_edge("C","E",capacity=6)
J.add_edge("D","E",capacity=3)
J.add_edge("D","F",capacity=6)
J.add_edge("E","F",capacity=9)
nx.draw_networkx(J)
```

ahora si usamos los comandos de la libreria


```python

valor_flujo, flujo_dict = nx.maximum_flow(J, 'A', 'F') # Este comando nos indica cuál es el valor numérico de nuestro flujo máximo además decirnos cuál es el flujo en cada arco
print(f"Valor del flujo máximo: {valor_flujo}")
print("Flujo en cada arista:")
print(flujo_dict)
```

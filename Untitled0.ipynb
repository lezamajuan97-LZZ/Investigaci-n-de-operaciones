# **Mis notas (investigacion de operaciones)**


primera red

```python
import networkx as nx #libreria para el estudio de redes
```

```python
G=nx.DiGraph()
G.add_edge(1,2,capacity=2)
G.add_edge(1,3,capacity=2)
G.add_edge(2,4,capacity=2)
G.add_edge(3,4,capacity=2)
G.add_edge(2,3,capacity=2)
nx.draw_networkx(G)
```

Esta es otra forma De hacer La red

```python
G=nx.DiGraph()
a=(1,2)
b=(1,3)
c=(2,4)
d=(3,4)
e=(2,3)
G.add_edge(*a)
G.add_edge(*b)
G.add_edge(*c)
G.add_edge(*d)
G.add_edge(*e)
nx.draw_networkx(G)
```

Siguiente código muestra otra forma de hacer la red y además cómo se devuelve el número de nodos y de arcos

```python
G=nx.DiGraph()
G.add_edges_from([(1, 2), (1, 3), (2,4),(3,4), (2,3)])
nx.draw_networkx(G)
n= G.number_of_nodes()
print("esta red consta de", n,"nodos")
m = G.number_of_edges()
print("esta red consta de", m,"arcos")
```
